```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"D:\VISUAL STUDIO\python programming\loan_approval.csv")
```

```python
df.info()
```

```python
df.drop(['name','city'],axis= 1,inplace=True)
```

```python
df
```

```python
df.isna().sum()
```

```python
df['loan_approved'].unique()
```

```python
loan_app_dict = {
    False : 0,  True:1
}
```

```python

df['loan_approved'] = df['loan_approved'].map(loan_app_dict)
```

```python
df
```

```python
sns.countplot(x='loan_approved',data=df)
plt.title("Loan Approval Distribution")
plt.show()
```

```python
sns.barplot(x='loan_approved', y='credit_score',data=df)
plt.title("Credit Score vs Loan Approved")
plt.show()
```

```python
sns.barplot(x='loan_approved',y='income',data=df)
plt.title("Income vs Loan Approval")
plt.show()
```

```python
sns.boxplot(x='loan_approved',y= 'loan_amount',data=df)
plt.title("Loan Amount vs Loan Approval")
plt.show()
```

```python
sns.boxplot(x='loan_approved',y='years_employed',data=df)
plt.title('Employement Years vs Loan Approval')
plt.show()
```

```python
sns.scatterplot(x='income',y='credit_score',data=df,hue='loan_approved')
plt.title("Income vs Credit Score")
plt.show()
```
